```python
import os
import random
import shutil
```

```python
BASE_DIR = os.getcwd()
```

```python
def makedir(new_dir):
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
```

```python
dataset_dir=os.path.join(BASE_DIR,"RMB", "RMB_data")
split_dir=os.path.join(BASE_DIR,"RMB", "rmb_split")
train_dir = os.path.join(split_dir, "train")
valid_dir = os.path.join(split_dir, "valid")
test_dir = os.path.join(split_dir, "test")
```

```python
train_pct = 0.8
valid_pct = 0.1
test_pct = 0.1
```

```python
raws=os.listdir(os.path.join(dataset_dir, "100"))
imgs = list(filter(lambda x: x.endswith('.jpg'), raws))
img = imgs[1]
src = os.path.join(dataset_dir, "100", img)
dst = os.path.join(train_dir, "100")
makedir(dst)
```

```python
for dirpath, dirnames, filenames in os.walk(dataset_dir):
    # print(f'Current directory: {dirpath}')
    # print(f'Subdirectories: {dirnames}')
    # print(f'Files: {filenames}')
    for sub_dir in dirnames:
        imgs = os.listdir(os.path.join(dirpath, sub_dir))
        imgs = list(filter(lambda x: x.endswith('.jpg'), imgs))
        random.shuffle(imgs)
        img_count = len(imgs)
        
        train_point = int(img_count * train_pct)
        valid_point = int(img_count * (train_pct + valid_pct))
        
        for i in range(img_count):
            img = imgs[i]
            src = os.path.join(dataset_dir, sub_dir, img)
            if i < train_point:
                dst = os.path.join(train_dir, sub_dir)
            elif i < valid_point:
                dst = os.path.join(valid_dir, sub_dir)
            else:
                dst = os.path.join(test_dir, sub_dir)
            makedir(dst)
            shutil.copy(src, dst)
            print(f'{src} -> {dst}')
    


```
